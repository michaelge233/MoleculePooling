```python
from rdkit.Chem import MolToSmiles, SDMolSupplier,RemoveHs

suppl=SDMolSupplier('./gdb9.sdf',sanitize = True,removeHs = True)
error_i=[]
smiles_list=[]
for i in range(len(suppl)):
    try:
        smiles_list.append(MolToSmiles(suppl[i]))
    except:
        error_i.append(i)
        smiles_list.append("")
```

```python
from rdkit.Chem import MolFromSmiles
len(error_i)
```

```python
with open("./gdb9.sdf.csv","r") as fo:
    file_list=fo.readlines()
file_list[0]="smiles,"+file_list[0]
for i in range(1,len(file_list)):
    file_list[i]=smiles_list[i-1]+","+file_list[i]
with open("./QM9.csv","w") as fo:
    for i in range(len(file_list)):
        if i-1 not in error_i:
            fo.write(file_list[i])
```
