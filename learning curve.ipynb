# QM9

```python
import numpy as np
import pandas as pd

def get_meanstd(table_list):
    min_list=[]
    for table in table_list:
        min_i=table["val"].values.argmin()
        min_list.append(table["test"][min_i])
    min_list=np.array(min_list)
    return np.mean(min_list),np.std(min_list)


add1=pd.read_csv("./result/add_QM9_1.csv")
add2=pd.read_csv("./result/add_QM9_2.csv")
add3=pd.read_csv("./result/add_QM9_3.csv")
mean1=pd.read_csv("./result/mean_QM9_1.csv")
mean2=pd.read_csv("./result/mean_QM9_2.csv")
mean3=pd.read_csv("./result/mean_QM9_3.csv")
ring1=pd.read_csv("./result/RING_QM9_1.csv")
ring2=pd.read_csv("./result/RING_QM9_2.csv")
ring3=pd.read_csv("./result/RING_QM9_3.csv")
diff1=pd.read_csv("./result/DIFF_QM9_1.csv")
diff2=pd.read_csv("./result/DIFF_QM9_2.csv")
diff3=pd.read_csv("./result/DIFF_QM9_3.csv")
```

```python
get_meanstd([add1,add2,add3])
```

```python
get_meanstd([mean1,mean2,mean3])
```

```python
get_meanstd([ring1,ring2,ring3])
```

```python
get_meanstd([diff1,diff2,diff3])
```

```python
import matplotlib.pyplot as plt

add_train=(add1["train"]+add2["train"]+add3["train"])/3
mean_train=(mean1["train"].iloc[:100]+mean2["train"]+mean3["train"])/3
ring_train=(ring1["train"]+ring2["train"]+ring3["train"])/3
diff_train=(diff1["train"]+diff2["train"]+diff3["train"])/3

plt.plot(np.arange(add_train.shape[0]),add_train)
plt.plot(np.arange(mean_train.shape[0]),mean_train)
plt.plot(np.arange(ring_train.shape[0]),ring_train)
plt.plot(np.arange(diff_train.shape[0]),diff_train)

plt.ylim(0.006,0.03)
plt.yscale("log")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

add_val=(add1["val"]+add2["val"]+add3["val"])/3
mean_val=(mean1["val"]+mean2["val"]+mean3["val"])/3
ring_val=(ring1["val"]+ring2["val"]+ring3["val"])/3
diff_val=(diff1["val"]+diff2["val"]+diff3["val"])/3

plt.plot(np.arange(add_val.shape[0])+1,add_val,label="Sum pool")
plt.plot(np.arange(mean_val.shape[0])+1,mean_val,label="Mean pool")
plt.plot(np.arange(ring_val.shape[0])+1,ring_val,label="RingPool")
plt.plot(np.arange(diff_val.shape[0])+1,diff_val,label="DiffPool")

plt.yscale("log")
plt.ylim(0.007,0.02)
plt.xlim(0,100)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Epoch",fontsize=24)
plt.ylabel("Validation MAE (Hatree)",fontsize=24)
plt.legend(fontsize=20)

bwidth=2
plt.tick_params(which="major",width=bwidth,length=5)
plt.tick_params(which="minor",width=bwidth,length=3)
tk=plt.gca()
tk.spines["bottom"].set_linewidth(bwidth)
tk.spines["top"].set_linewidth(bwidth)
tk.spines["left"].set_linewidth(bwidth)
tk.spines["right"].set_linewidth(bwidth)
plt.show()
```
